### df_northern_land in son hali

```python
df_northern_land['hemisphere'] = 'north'
df_northern_land.rename(columns={'Anomaly': 'AnomalyLand'}, inplace=True)

df_northern_land.head()
```

```python
df_northern_ocean = pd.read_csv("data_northern_ocean_temp.csv")
df_northern_ocean['Year'] = pd.to_datetime(df_northern_ocean['Year'], format='%Y%m')
df_northern_ocean = df_northern_ocean.groupby(df_northern_ocean['Year'].dt.year)['Anomaly'].mean().reset_index()

```

#### df_northern_ocean ın son hali

```python
df_northern_ocean['hemisphere'] = 'north'
df_northern_ocean.rename(columns={'Anomaly': 'AnomalyOcean'}, inplace=True)

df_northern_ocean.head()
```

#### df_southern_land in son hali

```python
df_southern_land = pd.read_csv("data_southern_land_temp.csv")
df_southern_land['Year'] = pd.to_datetime(df_southern_land['Year'], format='%Y%m')
df_southern_land = df_southern_land.groupby(df_southern_land['Year'].dt.year)['Anomaly'].mean().reset_index()
df_southern_land['hemisphere'] = 'south'
df_southern_land.rename(columns={'Anomaly': 'AnomalyLand'}, inplace=True)

df_southern_land.head()
```

#### df_southern_ocean ın son hali

```python
df_southern_ocean = pd.read_csv("data_southern_ocean_temp.csv")
df_southern_ocean['Year'] = pd.to_datetime(df_southern_ocean['Year'], format='%Y%m')
df_southern_ocean = df_southern_ocean.groupby(df_southern_ocean['Year'].dt.year)['Anomaly'].mean().reset_index()
df_southern_ocean['hemisphere'] = 'south'
df_southern_ocean.rename(columns={'Anomaly': 'AnomalyOcean'}, inplace=True)

df_southern_ocean.head()
```

```python
df_land_temp = pd.concat([df_northern_land, df_southern_land], ignore_index=True)
df_land_temp.head()
```

```python
df_ocean_temp = pd.concat([df_northern_ocean, df_southern_ocean], ignore_index=True)
df_ocean_temp.head()
```

```python
df_co= pd.read_csv("CO2_global_CEDS_emissions_by_fuel_2021_04_21.csv")
```

```python
df_co.head()
```

```python
df_melted = df_co.melt(id_vars=['fuel', 'em', 'units'], var_name='Year', value_name='Value')

# year sıralayalım
df_melted['Year'] = df_melted['Year'].str.extract('X(\d+)', expand=False).astype(int)

#transpozunu alacağım
df_transposed = df_melted.pivot(index='Year', columns='fuel', values='Value')

# year tekrar eklensin
df_co_res = df_transposed.reset_index().rename_axis(columns=None)

print(df_co_res)
```

```python
df_melted = df_co.melt(id_vars=['fuel', 'em', 'units'], var_name='Year', value_name='Value')

# year sıralayalım
df_melted['Year'] = df_melted['Year'].str.extract('X(\d+)', expand=False).astype(int)

#transpozunu alacağım
df_transposed = df_melted.pivot(index='Year', columns='fuel', values='Value')

# year tekrar eklensin
df_co_res_s = df_transposed.reset_index().rename_axis(columns=None)
df_co_res_s['hemisphere'] ='south'

print(df_co_res_s)
```
